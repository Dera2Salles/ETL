```python
mockDataBase = { "Alice" : (19 , ["Rust" , "Golang" , 'C']) , "Rado" : (23 , ["Rust" , "Python" ]) , 'Koto' : (40  , ["Python" , "JS" , "VB"])}

def optimized(mock) :
  listSkills = []
  result = {}

  for key in mock.keys():
    for value in mock[key][1]:
      listSkills.append(value)

  for value in set(listSkills):
    result[value] = listSkills.count(value)

  return result
```

```python
optimized(mockDataBase)
```

```python
listParameters = [2, 30 , 150 , 180 , 7 , 111 , 20,8,113,330,1950]

def sumThreeAndFiveDivisible(listParamaters):
    result = 0
    for value in listParamaters:
        if ((value % 3== 0) and (value % 5== 0)  ) :
            result += value
    return result
        
sumThreeAndFiveDivisible(listParameters)
```

```python
myList = [2, 30 , 150 , 180 , 7 , 111 , 20,8,113,330,1950]

def findMaxValueIndexOfSevenDivisible(listParameters):
    resteList = []
    maxValueIndex = []
    duplicateValue = []
    for value in listParameters:
        resteList.append(value % 7)
        
    for index, value in enumerate(resteList):
        if(resteList.count(value) > 1):
            duplicateValue.append(resteList[index])

    for index , value in enumerate(resteList):
        if value == max(resteList):
            maxValueIndex.append(index)
                
    return maxValueIndex

findMaxValueIndexOfSevenDivisible(myList)
```
